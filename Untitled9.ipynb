```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, MaxPooling1D, Flatten, Dense, Dropout, BatchNormalization

# Load CSV
df = pd.read_csv('UNSW-NB15_1.csv')

# Randomly select 1500 normal and 1500 attack
df_attack = df[df['Label'] == 1].sample(n=1500, random_state=42)
df_normal = df[df['Label'] == 0].sample(n=1500, random_state=42)
df_balanced = pd.concat([df_attack, df_normal]).sample(frac=1, random_state=42).reset_index(drop=True)

# Drop unwanted columns
drop_cols = ['srcip', 'dstip', 'Stime', 'Ltime', 'attack_cat']  # attack_cat: drop for binary
df_balanced = df_balanced.drop(columns=drop_cols)

# Identify categorical columns
categorical_cols = ['proto', 'state', 'service']
numerical_cols = [col for col in df_balanced.columns if col not in categorical_cols + ['Label']]

# One-hot encode categoricals
ohe = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
cat_encoded = ohe.fit_transform(df_balanced[categorical_cols])

# Standardize numerical columns
scaler = StandardScaler()
num_scaled = scaler.fit_transform(df_balanced[numerical_cols])

# Merge features
X = np.hstack([num_scaled, cat_encoded])
y = df_balanced['Label'].values

# Expand dims for CNN [samples, features, 1]
X = np.expand_dims(X, axis=2)

# Train/test split
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.2, random_state=42)

# CNN Model
input_shape = (X_train.shape[1], 1)
model = Sequential([
    Conv1D(32, 3, activation='relu', input_shape=input_shape),
    BatchNormalization(),
    MaxPooling1D(2),
    Conv1D(64, 3, activation='relu'),
    BatchNormalization(),
    MaxPooling1D(2),
    Flatten(),
    Dense(64, activation='relu'),
    Dropout(0.3),
    Dense(1, activation='sigmoid')
])
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.summary()

# Train model
history = model.fit(
    X_train, y_train,
    validation_data=(X_test, y_test),
    epochs=20,
    batch_size=64,
    verbose=2
)

# Evaluate
loss, acc = model.evaluate(X_test, y_test, verbose=0)
print(f"Test Accuracy: {acc:.4f}")
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from tensorflow.keras.models import Sequential, clone_model
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization
from tensorflow.keras.optimizers import Adam

# 1. Load and preprocess data (as in your code)
df = pd.read_csv('UNSW-NB15_1.csv')
df.columns = df.columns.str.strip()

df_attack = df[df['Label'] == 1].sample(n=1500, random_state=42)
df_normal = df[df['Label'] == 0].sample(n=1500, random_state=42)
df_balanced = pd.concat([df_attack, df_normal]).sample(frac=1, random_state=42).reset_index(drop=True)
drop_cols = ['srcip', 'dstip', 'Stime', 'Ltime', 'attack_cat']
df_balanced = df_balanced.drop(columns=drop_cols)

categorical_cols = ['proto', 'state', 'service']
numerical_cols = [
    col for col in df_balanced.columns
    if col not in categorical_cols + ['Label']
    and pd.api.types.is_numeric_dtype(df_balanced[col])
]
ohe = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
cat_encoded = ohe.fit_transform(df_balanced[categorical_cols])
num_data = df_balanced[numerical_cols].apply(pd.to_numeric, errors='coerce').fillna(0)
scaler = StandardScaler()
num_scaled = scaler.fit_transform(num_data)
X = np.hstack([num_scaled, cat_encoded])
y = df_balanced['Label'].values.astype(np.float32)

# 2. Split into train/test (80/20), then split train set among 5 clients
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.2, random_state=42)
client_data = []
num_clients = 5
data_per_client = len(X_train) // num_clients

for i in range(num_clients):
    start = i * data_per_client
    end = (i + 1) * data_per_client if i < num_clients - 1 else len(X_train)
    client_X = X_train[start:end]
    client_y = y_train[start:end]
    client_data.append((client_X, client_y))
```

```python
# 3. Model definition function
def build_model(input_dim):
    model = Sequential([
        Dense(128, activation='relu', input_dim=input_dim),
        BatchNormalization(),
        Dropout(0.4),
        Dense(64, activation='relu'),
        BatchNormalization(),
        Dropout(0.3),
        Dense(32, activation='relu'),
        Dense(1, activation='sigmoid')
    ])
    optimizer = Adam(learning_rate=0.0001)
    model.compile(optimizer=optimizer, loss='binary_crossentropy', metrics=['accuracy'])
    return model

# 4. Federated learning parameters
global_rounds = 50    # Number of global rounds
local_epochs = 5     # Number of epochs per client per global round
batch_size = 64
```

```python
# 5. Initialize global model and clients
input_dim = X_train.shape[1]
global_model = build_model(input_dim)
client_models = [build_model(input_dim) for _ in range(num_clients)]

# Initialize clients' models with global weights
global_weights = global_model.get_weights()
for model in client_models:
    model.set_weights(global_weights)

# 6. Federated training loop
for rnd in range(global_rounds):
    print(f"\n--- Global Round {rnd + 1} ---")
    local_weights = []
    # Each client trains locally
    for i, (model, (X_c, y_c)) in enumerate(zip(client_models, client_data)):
        model.set_weights(global_weights)  # Sync with global
        model.fit(X_c, y_c, epochs=local_epochs, batch_size=batch_size, verbose=0)
        local_weights.append(model.get_weights())
        print(f"Client {i + 1} local training done.")

    # Aggregate: Federated averaging (simple mean)
    new_weights = []
    for weights in zip(*local_weights):
        new_weights.append(np.mean(weights, axis=0))
    global_weights = new_weights

    # Update all models with the aggregated weights
    global_model.set_weights(global_weights)
    for model in client_models:
        model.set_weights(global_weights)

    # Evaluate global model on test set
    loss, acc = global_model.evaluate(X_test, y_test, verbose=0)
    print(f"Global Model Test Accuracy after round {rnd + 1}: {acc:.4f}")

# 7. Final evaluation
loss, acc = global_model.evaluate(X_test, y_test, verbose=0)
print(f"\nFinal Global Model Test Accuracy: {acc:.4f}")
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, StandardScaler, LabelEncoder
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.utils import to_categorical
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Load data
df = pd.read_csv('UNSW-NB_new.csv')
df.columns = df.columns.str.strip()

# 2. Balance classes (sample min_count from each class)
min_count = df['Label'].value_counts().min()
balanced_df = pd.concat([
    df[df['Label'] == label].sample(n=min_count, random_state=42)
    for label in df['Label'].unique()
]).sample(frac=1, random_state=42).reset_index(drop=True)

# 3. Drop columns not needed for learning
drop_cols = ['srcip', 'dstip', 'Stime', 'Ltime', 'attack_cat']
balanced_df = balanced_df.drop(columns=[col for col in drop_cols if col in balanced_df.columns])

# 4. Define categorical/numerical columns
categorical_cols = ['proto', 'state', 'service']
numerical_cols = [
    col for col in balanced_df.columns
    if col not in categorical_cols + ['Label']
    and pd.api.types.is_numeric_dtype(balanced_df[col])
]

# 5. Encode features
ohe = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
cat_encoded = ohe.fit_transform(balanced_df[categorical_cols])
num_data = balanced_df[numerical_cols].apply(pd.to_numeric, errors='coerce').fillna(0)
scaler = StandardScaler()
num_scaled = scaler.fit_transform(num_data)
X = np.hstack([num_scaled, cat_encoded])

# 6. Encode labels (integer → one-hot)
le = LabelEncoder()
y_int = le.fit_transform(balanced_df['Label'])
num_classes = len(le.classes_)
y = to_categorical(y_int, num_classes=num_classes)

# 7. Split train/test, then train split among clients
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y_int, test_size=0.2, random_state=42)
client_data = []
num_clients = 5
data_per_client = len(X_train) // num_clients

for i in range(num_clients):
    start = i * data_per_client
    end = (i + 1) * data_per_client if i < num_clients - 1 else len(X_train)
    client_X = X_train[start:end]
    client_y = y_train[start:end]
    client_data.append((client_X, client_y))

# 8. Build multiclass model
def build_model(input_dim, num_classes):
    model = Sequential([
        Dense(128, activation='relu', input_dim=input_dim),
        BatchNormalization(),
        Dropout(0.4),
        Dense(64, activation='relu'),
        BatchNormalization(),
        Dropout(0.3),
        Dense(32, activation='relu'),
        Dense(num_classes, activation='softmax')
    ])
    optimizer = Adam(learning_rate=0.0001)
    model.compile(optimizer=optimizer, loss='categorical_crossentropy', metrics=['accuracy'])
    return model

# 9. Federated parameters
global_rounds = 10
local_epochs = 10
batch_size = 64
input_dim = X_train.shape[1]

# 10. Init models
global_model = build_model(input_dim, num_classes)
client_models = [build_model(input_dim, num_classes) for _ in range(num_clients)]
global_weights = global_model.get_weights()
for model in client_models:
    model.set_weights(global_weights)

# 11. Federated learning loop
for rnd in range(global_rounds):
    print(f"\n--- Global Round {rnd + 1} ---")
    local_weights = []
    for i, (model, (X_c, y_c)) in enumerate(zip(client_models, client_data)):
        model.set_weights(global_weights)
        model.fit(X_c, y_c, epochs=local_epochs, batch_size=batch_size, verbose=0)
        local_weights.append(model.get_weights())
        print(f"Client {i + 1} local training done.")

    # Aggregate
    new_weights = []
    for weights in zip(*local_weights):
        new_weights.append(np.mean(weights, axis=0))
    global_weights = new_weights

    # Sync
    global_model.set_weights(global_weights)
    for model in client_models:
        model.set_weights(global_weights)

    # Evaluate global model
    loss, acc = global_model.evaluate(X_test, y_test, verbose=0)
    print(f"Global Model Test Accuracy after round {rnd + 1}: {acc:.4f}")

# Predict class probabilities
y_pred = global_model.predict(X_test)
# Get the class with highest probability for each sample
y_pred_label = np.argmax(y_pred, axis=1)
y_true_label = np.argmax(y_test, axis=1)

# Classification report (summary)
print("\nClassification Report:")
print(classification_report(y_true_label, y_pred_label, target_names=[str(cls) for cls in le.classes_]))

# Confusion Matrix
cm = confusion_matrix(y_true_label, y_pred_label)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=[str(cls) for cls in le.classes_],
            yticklabels=[str(cls) for cls in le.classes_])
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.tight_layout()
plt.show()

# Final evaluation
loss, acc = global_model.evaluate(X_test, y_test, verbose=0)
print(f"\nFinal Global Model Test Accuracy: {acc:.4f}")

# Optional: Class-wise report
from sklearn.metrics import classification_report
y_pred = global_model.predict(X_test)
y_pred_label = np.argmax(y_pred, axis=1)
y_true_label = np.argmax(y_test, axis=1)
print(classification_report(y_true_label, y_pred_label, target_names=[str(cls) for cls in le.classes_]))
```

# NON IID MULTICLASSS DISTRIBUTION


```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, StandardScaler, LabelEncoder
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.utils import to_categorical
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# Load and clean data
df = pd.read_csv('UNSW-NB_new.csv')
df.columns = df.columns.str.strip()
min_count = df['Label'].value_counts().min()
balanced_df = pd.concat([
    df[df['Label'] == label].sample(n=min_count, random_state=42)
    for label in df['Label'].unique()
]).sample(frac=1, random_state=42).reset_index(drop=True)

drop_cols = ['srcip', 'dstip', 'Stime', 'Ltime', 'attack_cat']
balanced_df = balanced_df.drop(columns=[col for col in drop_cols if col in balanced_df.columns])

categorical_cols = ['proto', 'state', 'service']
numerical_cols = [
    col for col in balanced_df.columns
    if col not in categorical_cols + ['Label']
    and pd.api.types.is_numeric_dtype(balanced_df[col])
]

ohe = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
cat_encoded = ohe.fit_transform(balanced_df[categorical_cols])
num_data = balanced_df[numerical_cols].apply(pd.to_numeric, errors='coerce').fillna(0)
scaler = StandardScaler()
num_scaled = scaler.fit_transform(num_data)
X = np.hstack([num_scaled, cat_encoded])

le = LabelEncoder()
y_int = le.fit_transform(balanced_df['Label'])
num_classes = len(le.classes_)
y = to_categorical(y_int, num_classes=num_classes)

# Train/test split
X_train, X_test, y_train, y_test, y_train_int, y_test_int = train_test_split(
    X, y, y_int, stratify=y_int, test_size=0.2, random_state=42)
```

# Non-IID Uneven Client Split

```python
# Assumes X_train, y_train, y_train_int are already defined
import numpy as np

all_classes = [0, 1, 2, 3, 4]

# Now, each client sees 60% dominant (3 classes), 40% rest
dominant_class_map = [
    [0, 1, 2], [1, 2, 3], [2, 3, 4], [3, 4, 0], [4, 0, 1],
    [0, 2, 3], [1, 3, 4], [2, 4, 0], [3, 0, 1], [4, 1, 2]
]

client_data = []
rng = np.random.default_rng(42)
used_indices = set()

for client_idx in range(10):
    dom_classes = dominant_class_map[client_idx]
    client_size = rng.integers(3000, 5001)  # min 3000, max 5000
    dom_size = int(client_size * 0.6)
    rest_size = client_size - dom_size

    available_indices = np.setdiff1d(np.arange(len(X_train)), list(used_indices))
    dom_pool = [i for i in available_indices if y_train_int[i] in dom_classes]
    rest_classes = [c for c in all_classes if c not in dom_classes]
    rest_pool = [i for i in available_indices if y_train_int[i] in rest_classes]

    dom_sample = rng.choice(dom_pool, size=dom_size, replace=len(dom_pool) < dom_size)
    rest_sample = rng.choice(rest_pool, size=rest_size, replace=len(rest_pool) < rest_size)
    indices = np.concatenate([dom_sample, rest_sample])
    # used_indices.update(indices)  # Uncomment if you want disjoint clients

    client_X = X_train[indices]
    client_y = y_train[indices]
    client_data.append((client_X, client_y, len(indices)))
    # Class histogram
    labs = np.argmax(client_y, axis=1)
    unique, counts = np.unique(labs, return_counts=True)
    print(f"Client {client_idx+1}: total={len(indices)}, dominant={dom_classes}, class_dist={dict(zip(unique, counts))}")

```

## DNN Model Definition

```python
def build_model(input_dim, num_classes):
    model = Sequential([
        Dense(128, activation='relu', input_dim=input_dim),
        BatchNormalization(),
        Dropout(0.4),
        Dense(64, activation='relu'),
        BatchNormalization(),
        Dropout(0.3),
        Dense(32, activation='relu'),
        Dense(num_classes, activation='softmax')
    ])
    optimizer = Adam(learning_rate=0.0001)
    model.compile(optimizer=optimizer, loss='categorical_crossentropy', metrics=['accuracy'])
    return model
```

# Federated Learning with Weighted Aggregation

```python
global_rounds = 50
local_epochs = 10
batch_size = 64
input_dim = X_train.shape[1]

global_model = build_model(input_dim, num_classes)
client_models = [build_model(input_dim, num_classes) for _ in range(10)]
global_weights = global_model.get_weights()
for model in client_models:
    model.set_weights(global_weights)

for rnd in range(global_rounds):
    print(f"\n--- Global Round {rnd + 1} ---")
    local_weights = []
    local_sizes = []
    for i, (model, (X_c, y_c, n_c)) in enumerate(zip(client_models, client_data)):
        model.set_weights(global_weights)
        model.fit(X_c, y_c, epochs=local_epochs, batch_size=batch_size, verbose=0)
        local_weights.append(model.get_weights())
        local_sizes.append(n_c)
        print(f"Client {i + 1} local training done (n={n_c}).")

    # Weighted aggregation
    total_size = sum(local_sizes)
    new_weights = []
    for weights in zip(*local_weights):
        new_w = sum(w * n for w, n in zip(weights, local_sizes)) / total_size
        new_weights.append(new_w)
    global_weights = new_weights
    global_model.set_weights(global_weights)
    for model in client_models:
        model.set_weights(global_weights)
    loss, acc = global_model.evaluate(X_test, y_test, verbose=0)
    print(f"Global Model Test Accuracy after round {rnd + 1}: {acc:.4f}")
```

# Evaluation – Classification Report and Confusion Matrix

```python
y_pred = global_model.predict(X_test)
y_pred_label = np.argmax(y_pred, axis=1)
y_true_label = np.argmax(y_test, axis=1)

print("\nClassification Report:")
print(classification_report(y_true_label, y_pred_label, target_names=[str(cls) for cls in le.classes_]))

cm = confusion_matrix(y_true_label, y_pred_label)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=[str(cls) for cls in le.classes_],
            yticklabels=[str(cls) for cls in le.classes_])
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.tight_layout()
plt.show()

loss, acc = global_model.evaluate(X_test, y_test, verbose=0)
print(f"\nFinal Global Model Test Accuracy: {acc:.4f}")
```
